```python
import pandas as pd
df=pd.read_csv("academy_awards.csv",encoding="ISO-8859-1")
df

```

```python
###Exploring the data. Looking at various column data types, shapes and unique values###

print(df.shape)
print(df.columns)
```

```python
print (df["Unnamed: 5"].unique())
print (df["Unnamed: 6"].unique())
print (df["Unnamed: 7"].unique())
print (df["Unnamed: 8"].unique())
print (df["Unnamed: 9"].unique())
print (df["Unnamed: 10"].unique())

```

```python
###Exploring Data ###
df["Additional Info"].unique().tolist()



```

```python
###Cleaning the year column

df["Year"]=df['Year'].str[0:4]
df["Year"]=df["Year"].astype("int64")
later_than_2000=df[df["Year"]>2000]






    



        
```

```python
### Choosing only the relevant categories of awars

li=df["Category"].unique().tolist()[0:4]
later_than_2000=later_than_2000[later_than_2000["Category"].isin(li)]
later_than_2000["Category"].unique()



```

```python
###Converting Won? Column to a boolean

later_than_2000["Won"]=later_than_2000["Won?"].replace(['YES','NO'],[1,0])

```

```python
### Deleting extra Columns

final_nominations=later_than_2000.drop(['Won?','Unnamed: 5','Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10'],axis=1)
final_nominations
```

```python
additional_info_one=final_nominations["Additional Info"].str.strip("'}")
additional_info_two=additional_info_one.str.split(" {'")
final_nominations["Movie"]=additional_info_two.str[0]
final_nominations["Character"]=additional_info_two.str[1]
final_nominations=final_nominations.drop('Additional Info',axis=1)
final_nominations
```

```python
## Export To Sqlite 
import sqlite3
conn=sqlite3.connect("nominations.db")
final_nominations.to_sql("nominations",conn, index=False)
```
